```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Load dataset
data = pd.read_excel(r"C:\Users\god\Desktop\patients_stats_project.xlsx")

# 1️⃣ Descriptive Statistics
print("Mean Recovery Days:", data['Recovery_Days'].mean())
print("Median Recovery Days:", data['Recovery_Days'].median())
print("Mode Recovery Days:", data['Recovery_Days'].mode()[0])
print("Variance:", data['Recovery_Days'].var())
print("Standard Deviation:", data['Recovery_Days'].std())

# 2️⃣ Correlation: Age vs Recovery Days
print("\nCorrelation between Age and Recovery Days:")
print(data[['Age', 'Recovery_Days']].corr())

# 3️⃣ T-Test: Compare Treatment A vs Treatment B
A = data[data['Treatment_Type'] == 'A']['Recovery_Days']
B = data[data['Treatment_Type'] == 'B']['Recovery_Days']

t_stat, p_value = stats.ttest_ind(A, B)
print("\nT-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("✅ Significant difference: Treatment B works better.")
else:
    print("❌ No significant difference between treatments.")

# 4️⃣ Visualization
plt.figure(figsize=(6,4))
data.boxplot(column='Recovery_Days', by='Treatment_Type', grid=False)
plt.title("Treatment Type vs Recovery Days")
plt.suptitle("")
plt.xlabel("Treatment Type")
plt.ylabel("Recovery Days")
plt.show()

# 5️⃣ Summary
print("\nProject Summary:")
print(f"- Average recovery time: {round(data['Recovery_Days'].mean(), 2)} days")
print(f"- Treatment A Avg: {round(A.mean(),2)}, Treatment B Avg: {round(B.mean(),2)}")
print("✅ Analysis Completed.")
```
